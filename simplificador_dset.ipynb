```python

new_base_path = 'D:/normal_small'

# Criar estrutura de pastas em D:
for pasta in os.listdir(base_path)[:4]:
    pasta_path = os.path.join(base_path, pasta)
    if os.path.isdir(pasta_path):
        new_pasta_path = os.path.join(new_base_path, pasta)
        os.makedirs(new_pasta_path, exist_ok=True)
        # Iterar sobre os 49 arquivos CSV em cada pasta
        for arquivo in os.listdir(pasta_path)[:49]:
            if arquivo.endswith('.csv'):
                file_path = os.path.join(pasta_path, arquivo)
                # Ler o arquivo
                df = pd.read_csv(file_path)
                # Verificar se há colunas suficientes
                if len(df.columns) > 7:
                    df_reconstruido = df.drop(df.columns[[1, 2, 3, 7]], axis=1)
                    # Salvar no mesmo índice de pasta com nome ajustado
                    output_filename = f'reconstruido_{os.path.basename(arquivo).replace(".csv", "_new.csv")}'
                    output_path = os.path.join(new_pasta_path, output_filename)
                    df_reconstruido.to_csv(output_path, index=False)
                    print(f"Salvo em: {output_path}")
                else:
                    print(f"Arquivo {arquivo} tem colunas insuficientes: {len(df.columns)}")
```

```python
import os
import pandas as pd

# Caminho base original e novo
base_path = r'C:\Files\mafaulda_bas_1\overhang\outer_race\35g'
new_base_path = 'D:/outer_race/35g'

# Criar pasta D:/nouter_race_small se não existir
os.makedirs(new_base_path, exist_ok=True)

# Verificar se a pasta base existe e listar arquivos
if not os.path.exists(base_path) or not os.path.isdir(base_path):
    print(f"Caminho {base_path} não encontrado ou não é uma pasta.")
else:
    arquivos = [f for f in os.listdir(base_path) if f.endswith('.csv')]
    if not arquivos:
        print(f"Nenhum arquivo .csv encontrado em {base_path}. Verifique o caminho.")
    else:
        print(f"Arquivos encontrados: {len(arquivos)}")
        # Iterar sobre os 49 arquivos CSV
        for arquivo in arquivos[:49]:
            file_path = os.path.join(base_path, arquivo)
            print(f"Lendo: {file_path}")
            try:
                df = pd.read_csv(file_path)
                print(f"Colunas encontradas: {len(df.columns)}")
                if len(df.columns) > 7:
                    df_reconstruido = df.drop(df.columns[[1, 2, 3, 7]], axis=1)
                    output_filename = f'reconstruido_{os.path.basename(arquivo).replace(".csv", "_new.csv")}'
                    output_path = os.path.join(new_base_path, output_filename)
                    df_reconstruido.to_csv(output_path, index=False)
                    print(f"Salvo em: {output_path}")
                else:
                    print(f"Arquivo {arquivo} tem colunas insuficientes: {len(df.columns)}")
            except Exception as e:
                print(f"Erro ao processar {arquivo}: {e}")
```
